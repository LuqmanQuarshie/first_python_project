### Replacing null values in each column with mean of each column and why?

```python
columns=['fixed acidity','citric acid','chlorides','free sulfur dioxide','total sulfur dioxide','density','pH','quality']
```

```python
for column in columns:
    data[column].fillna(data[column].mean(),inplace=True)
```

A popular strategy in data analysis with missing values is to replace null values with the column mean. Eliminating rows with missing values could lead to data loss and a smaller sample size, which would be bad for the analysis's accuracy and representativeness. The mean of the column can be used to replace null values in order to maintain the sample size and offer a reliable estimate of the missing data. Nevertheless, this method makes the assumption that the data is normally distributed and the missing values are randomly dispersed.This technique might not be appropriate for producing an accurate representation of the missing data if the missing values are not dispersed randomly or the data is not normally distributed. Other methods like interpolation, imputation, or removing rows and columns may be more suited in certain circumstances.

### Z-Scores and Outliers
The z-score indicates how many standard deviations a particular data point is distant from the mean of the distribution.

An outlier is a data point that lies far beyond the range of other data points. A popular technique for detecting outliers involves setting a threshold of 3 standard deviations.





```python
for column in data.columns:
    if column != 'ID' :
    #Calculating the z-scores for each data point of each column
       z_scores = (data[column] - data[column].mean()) / data[column].std()
    #output below is all z_scores values for each column
       print(z_scores)

     #outliers are identified using a threshold of 3 standard deviations
       outliers = data[abs(z_scores) > 3]
```

```python
outliers
```

```python
for column in outliers.columns:
    if column != 'ID':
        pl.title(f'Outliers in {column}')
        pl.xlabel('Range')
        
        sb.scatterplot(outliers[column])
        pl.show()
```
